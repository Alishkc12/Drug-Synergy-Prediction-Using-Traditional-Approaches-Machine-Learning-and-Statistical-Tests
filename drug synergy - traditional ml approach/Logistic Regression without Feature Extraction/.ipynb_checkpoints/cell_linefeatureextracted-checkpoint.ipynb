```python
X = drug_interaction.drop(columns=['label']).values  
y = drug_interaction['label'].values                 

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
from torch.utils.data import Dataset, DataLoader

class SynergyDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32).view(-1, 1, 651)  # reshape for CNN
        self.y = torch.tensor(y, dtype=torch.float32).view(-1, 1)       # binary labels

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = SynergyDataset(X_train, y_train)
val_dataset = SynergyDataset(X_val, y_val)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class DrugSynergyCNN(nn.Module):
    def __init__(self):
        super(DrugSynergyCNN, self).__init__()
        self.conv1 = nn.Conv1d(1, 16, kernel_size=5, padding=2)
        self.pool1 = nn.MaxPool1d(2)
        self.conv2 = nn.Conv1d(16, 32, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool1d(2)
        self.fc1 = nn.Linear(5184, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        x = F.relu(self.conv1(x))    # [B, 16, 384]
        x = self.pool1(x)            # [B, 16, 192]
        x = F.relu(self.conv2(x))    # [B, 32, 192]
        x = self.pool2(x)            # [B, 32, 96]
        x = x.view(x.size(0), -1)    # flatten
        x = F.relu(self.fc1(x))      # [B, 64]
        x = torch.sigmoid(self.fc2(x))  # [B, 1]
        return x

model = DrugSynergyCNN()
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training loop
for epoch in range(10):  # or more epochs
    model.train()
    total_loss = 0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss/len(train_loader):.4f}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import torch

def evaluate_model(model, dataloader, threshold=0.5):
    model.eval()
    y_true = []
    y_pred = []

    with torch.no_grad():
        for X_batch, y_batch in dataloader:
            outputs = model(X_batch)
            probs = torch.sigmoid(outputs)  # since it's binary classification
            preds = (probs > threshold).float()

            y_true.extend(y_batch.view(-1).tolist())
            y_pred.extend(preds.view(-1).tolist())

    # Convert to numpy just in case
    y_true = torch.tensor(y_true).numpy()
    y_pred = torch.tensor(y_pred).numpy()

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred)

    print("Confusion Matrix:\n", cm)
    print(f"Accuracy:  {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall:    {rec:.4f}")
    print(f"F1-score:  {f1:.4f}")
```

```python
evaluate_model(model, val_loader)
```
